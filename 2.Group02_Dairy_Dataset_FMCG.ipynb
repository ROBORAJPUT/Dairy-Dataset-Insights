# 1. Find total land area for each location.

```python
# Hint:.
#  Group by the "Location" column. Then, select the "Total Land Area (acres)" column within each group,
# and calculate the sum of values within each group.
to_grp = dairy_data.groupby("Location")["Total Land Area (acres)"].sum().reset_index()

# Display the resulting DataFrame 'to_grp'.
to_grp
```

# 2. Convert Location and Brand variables from dairy_data dataset into upper case.

```python
# Hint:.
# Use the applymap() function to transform specific columns in the 'dairy_data' DataFrame.
# For the 'Location' and 'Brand' columns, apply the lambda function to each cell.
# If the cell value is a string (str), convert it to uppercase using the upper() method.
# If the cell value is not a string (e.g., numeric or other data types), leave it unchanged.

dairy_data[['Location', 'Brand']] = dairy_data[['Location', 'Brand']].applymap(lambda x: x.upper() if isinstance(x, str) else x)

# Display the modified 'dairy_data' DataFrame.
dairy_data
```

# 3. How many unique brands are there, and what is the average price per unit for each brand?

```python
# Hint:.
# Get the unique values in the 'Brand' column and store them in the 'unique_Brand' variable.
unique_Brand = dairy_data['Brand'].unique()
unique_Brand

# Group the 'dairy_data' DataFrame by the 'Brand' column and calculate the mean of the 'Price per Unit' for each brand.
Brand_avg_prc= dairy_data.groupby('Brand')['Price per Unit'].mean().round(2)
Brand_avg_prc

# Print the number of unique brands by getting the length of the 'unique_Brand' array.
print("Number of Unique Brands:", len(unique_Brand))

#Print a message and the average price per unit for each brand.
print("Average Price per Unit for Each Brand:")
print(Brand_avg_prc)
```

# 5. Calculate the total price per unit of 'Sudha' brand milk products that are stored in 'Tetra Pack' containers.

# 6. Give the appropriate chart you've created to visualize the overall monthly Revenue trend?

```python
# Hint:.
# Create a new column 'Month_year' in the 'dairy_data' DataFrame by applying a lambda function to format the 'Production Date' column as '%Y-%m'.
dairy_data['Month_year'] = dairy_data['Production Date'].apply(lambda x : x.strftime('%Y-%m'))
dairy_data

# Group the 'dairy_data' DataFrame by the 'Month_year' column and calculate the sum of 'Approx. Total Revenue(INR)' for each month-year.
dd = dairy_data.groupby('Month_year')['Approx. Total Revenue(INR)'].sum().reset_index()
dd

# Create a plot with matplotlib to visualize the monthly revenue trends.
plt.figure(figsize=(16, 5))
plt.plot(dd['Month_year'], dd['Approx. Total Revenue(INR)'], marker = 'o', color='#0000FF')

# Customize the x-axis labels by rotating them vertically and adjusting the label size.
plt.xticks(rotation='vertical', size=8)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month_year")
plt.ylabel("Approx. Total Revenue(INR)")

# Show the plot.
plt.show()

# Why we use line chart:-  This type of data reveals trends, patterns, and seasonality, making it perfect for line charts.
```

# 10. How much year growth rate-Sales has been improved (CAGR) per location for all brand.

```python
# Hint:.
# Filter the data to select rows where the 'Year' is 2019.
loc_19=dairy_data[dairy_data['Year'] == 2019]

# Group the 2019 data by 'Location' and calculate the sum of 'Approx. Total Revenue(INR)' for each location.
gr_loc_19=loc_19.groupby('Location')['Approx. Total Revenue(INR)'].sum().reset_index()

# Filter the data to select rows where the 'Year' is 2022.
loc_22=dairy_data[dairy_data['Year'] == 2022]

# Group the 2022 data by 'Location' and calculate the sum of 'Approx. Total Revenue(INR)' for each location.
gr_loc_22=loc_22.groupby('Location')['Approx. Total Revenue(INR)'].sum().reset_index()

# Merge the 2019 and 2022 location data based on 'Location'.
gr_loc_19_22=pd.merge(gr_loc_19,gr_loc_22,how='inner',left_on='Location',right_on='Location')

# Calculate the Compound Annual Growth Rate (CAGR) for each location's revenue from 2019 to 2022.
gr_loc_19_22['CAGR']=(((gr_loc_19_22['Approx. Total Revenue(INR)_y']/gr_loc_19_22['Approx. Total Revenue(INR)_x'])**1/3)-1)*100

# Display the resulting DataFrame that contains location-wise revenue and CAGR information.
gr_loc_19_22.sort_values('CAGR', ascending = False).round(2)
```
